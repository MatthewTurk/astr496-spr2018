```python
import ipywidgets
```

```python
@ipywidgets.interact(lower = (0.0, 3.5), upper = (0.0, 3.5))
def show_image(lower=0.0, upper=3.5):
    plt.imshow(np.log10(image))
    plt.clim(lower, upper)
```
