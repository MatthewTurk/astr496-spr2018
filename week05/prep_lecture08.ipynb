```python
from PIL import Image
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.dpi"] = 100
```

```python
import numpy as np
```

```python
im_data = Image.open("/home/storage/downloads/wherespoochie.jpg")
im_array = np.array(im_data)
```

```python
import scipy.ndimage
```

```python
w = np.array([[1,1,1],[1,1,0],[1,0,0]])
```

```python
fw = np.array([w,w,w])
fw = np.array(3*[np.eye(3)])
fw = np.zeros((3,3,3))
fw = np.array([3 * np.array([[0,1,0],[0,1,0],[0,1,0]])])
```

```python
conv = scipy.ndimage.convolve(im_array, fw, mode='constant', cval=0.0)
```

```python
plt.imshow(conv)
```

```python
import ipywidgets
```

```python
@ipywidgets.interact(spike = (0.0, 100.0, .1))
def make_plot(spike):
    arr = np.ones((3,3))
    arr[1,1] = spike
    arr /= (arr*arr).sum()**0.5
    conv = scipy.ndimage.convolve(im_array[:,:,0], arr, mode='constant', cval = 0)
    plt.imshow(conv, cmap="gray")
```
