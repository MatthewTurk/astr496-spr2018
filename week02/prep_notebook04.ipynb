```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4], [5,6,7,8])

plt.savefig("images/img1.png")
```

```python
plt.plot([1,2,3,4], [5,6,7,8], color = 'b', linestyle = '--',                                             
            marker = 'x', linewidth = 2.0) 
plt.savefig("images/img2.png")
```

```python
plt.plot([1,2,3,4], [5,6,7,8], color = 'b', linestyle = '',                                             
            marker = 'o', markersize = 10.0) 
plt.savefig("images/img3.png")
```

```python
import numpy as np
xy = np.random.normal(size=(100000,2), loc = [0.5, 0.5])
plt.scatter(xy[:,0], xy[:,1])
plt.savefig("images/scatter1.png")
```

```python
import numpy as np
xy = np.random.normal(size=(100000,2), loc = [0.5, 0.5])
plt.hexbin(xy[:,0], xy[:,1])
plt.savefig("images/hexbin1.png")
```

```python
plt.hist(xy[:,0])
plt.savefig("images/hist1.png")
```
